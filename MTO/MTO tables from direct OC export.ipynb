```python
import pandas as pd
# from pandas import ExcelWriter
# from openpyxl import load_workbook
import numpy as np
# import csv
# from rdp import rdp
# from tqdm import tqdm
```

```python
fname = "testing case (1).xlsx"
fname = "Test2.xlsx"
writer = pd.ExcelWriter(fname+'.out.xlsx')
#set viewing parameters
pd.options.display.max_columns=40
pd.set_eng_float_format(accuracy=2, use_eng_prefix=True)
l_df=pd.read_excel(fname,sheet_name="lines",header=0)
l_df.head()
```

```python
l_df["Template name"].fillna(l_df["Name"], inplace=True)
l_df[l_df["Element type"]=="Line-Component"].head()
```

```python
column_order = ['Tot. count', 'Tot. weight (kg)', 'Tot. length (m)', '4 Com. Length (m)']
table = pd.pivot_table(l_df, values=column_order, 
                       index=['Element type','Template name', '1 Description'],
                       aggfunc={'Tot. count': np.sum, 
                                'Tot. weight (kg)': np.sum,
                                'Tot. length (m)': [min, max, np.sum],
                                '4 Com. Length (m)':[min, max, np.sum]
                               })

for manager in table.index.get_level_values(0).unique():
    temp_df = table.xs(manager, level=0)
    temp_df.to_excel(writer,manager+"pt")

# table2 = table.reindex(column_order, axis=1)
table
```
