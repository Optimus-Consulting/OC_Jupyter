## Seaborn Tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
# load some data
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.smoker.unique()
```

```python
sns.relplot(x="total_bill", y="tip", data=df, hue="sex")
```

```python
# Add a hue, like for each category, show the same graph with different color
sns.relplot(x="total_bill", y="tip", hue="smoker", data=df);
```

```python
# different style for each smoker category
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=df);
```

```python
# different style for each `time` category
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=df);
```

```python
#changing the hue to "size"
sns.relplot(x="total_bill", y="tip", hue="size", data=df);
```

```python
# color palette..
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=df);
```

```python
df.head(3)
```

```python
# controlling the size of the scatter plot circle size...()
# can make, bubble chart with this.
sns.relplot(x="total_bill", y="tip", size="size", hue="size", data=df);
```

```python
#  size="size", actually size data column
#  sizes = (min, max) size scales...
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=df);
```

```python
## Can be line plot, by changing the kind="line"
tmpdf = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=tmpdf)
g.fig.autofmt_xdate()
```

## Categorical Plot

```python
df.head()
```

```python
# catplot- x, y
sns.catplot(x="day", y="total_bill", data=df);
```

```python
# catplot- x, y with hue - size
sns.catplot(x="day", y="total_bill", data=df, hue="size");
```

```python
# catplot- x, y with hue - sex
sns.catplot(x="day", y="total_bill", data=df, hue="sex");
```

```python
# catplot- x, y with no jitter
sns.catplot(x="day", y="total_bill", jitter=False, data=df);
```

```python

sns.catplot(x="day", y="total_bill", kind="swarm", data=df);
```

```python

sns.catplot(x="day", y="total_bill", kind="strip", data=df);
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=df);
```

```python
df.head()
```

```python
# df.query("sex == 'Female' and tip > 1.5")
# df.query("tip > 1.5")
```

```python
# with condition
# pandas query
sns.catplot(x="size", y="total_bill", kind="swarm", data=df.query("size != 3"));
```

```python
# customize the order  order=["No", "Yes"]
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=df);
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=df);
```
