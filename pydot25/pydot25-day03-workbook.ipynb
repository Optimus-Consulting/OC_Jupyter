```python
with open(fname, encoding="utf-8") as f:
#     contents = f.read()
#     words = contents.split()
    total_words_count = 0
    word_dct = {}
    for line in f:
        words = line.split()
        total_words_count += len(words)
        for word in words:
            word_dct[word] = word_dct.get(word, 0) + 1

    #count = len(words)
    unique_count = len(word_dct)
    
    print(f"There are {total_words_count} words in total")
    print(f"There are Unique {unique_count} words in total")
```

```python
from collections import Counter
word_dct_counter=  Counter(word_dct)
# word_dct_counter
```

```python
for word in [ item[0] for item in word_dct_counter.most_common(6)]:
    del word_dct[word]
```

```python
src_fname = input("Enter source file")
target_fname = input("Enter target file")

tar_f = open(target_fname, mode='ab')

with open(src_fname, mode='rb') as src_f:
    while True:
        CHUK_SIZE = 1024
        content= src_f.read(CHUK_SIZE)
        if not content: break        
        tar_f.write(content)

tar_f.close()
```
