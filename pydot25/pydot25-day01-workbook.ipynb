```python
# Exercise 3
try:
    a = int(input("Enter the value of a:"))
    b = int(input("Enter the value of b:"))
    ops = input("Enter the operator: + - * /")

    if ops == "+":
        result = a + b
    elif ops == "-":
        result = a - b
    elif ops == "*":
        result = a * b
    else:
        result = "Invalid Operator"

    message = f'{a} {ops} {b} => : {result}'
    print( message )
except ValueError:
    print("Invalid values for a or b. Please enter ints..")
```

```python
# Exercise 3
try:
    a = int(input("Enter the value of a:"))
    b = int(input("Enter the value of b:"))
    ops = input("Enter the operator: + - * /")

    operation_str = f'{a} {ops} {b}'
    result = eval(operation_str)

    message = f'{a} {ops} {b} => : {result}'
    print( message )
except ValueError:
    print("Invalid values for a or b. Please enter ints..")
except Exception:
    print("something went wrong..")
```
