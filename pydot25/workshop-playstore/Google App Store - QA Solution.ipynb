```python
def func(x):
    if x < 2.5:
        return "2.5 and less"
    elif x < 3.5:
        return "3.5-2.5"
    elif x < 4.5:
        return "4.5-3.5"
    else:
        return "4.5-5.0"

df["Rating_cat"] = df.Rating.apply( func )
```

```python
res = df.groupby("Rating_cat")["App"].count()
res.sort_index().plot("bar")
```
