# Return crossing location, direction and lines name
From the list of lines check if they are crossing each other are retun crossing list in format:
{"id": number,
"position":[x,y],
"lines": [line1 name, line2 name],
"directions": [line1 direction at crossing, line2 direction at crossing]
}

```python
import pandas as pd
from shapely.geometry import Point, Polygon, LineString
```

```python
import json
with open('testing case.json', "r") as json_file:
    f=json_file.read()
    data = json.loads(f)
len(data)
```

```python
for i in data:
    print(i)
```

```python
wantedData=[]
for index, i in enumerate(data["elements"]):
    if i["type"]["name"]=="Line":
        name = i["name"]
        for j in i["properties"]:
            if j["name"]=="points":
                points =j["value"]
                wantedData.append([name,points])
                
df=pd.DataFrame(wantedData)
```

```python
df.columns=["Name", "Points"]
df
```

```python
lines=[]
for index, l in df.iterrows():
    line=LineString([Point((p["x"],p["y"])) for p in l.Points])
    lines.append(line)
df["line"]=lines
```

```python
df
```

### Jedno od rijesenja nikako optimalno da svaku liniju provjeriti sa svim ostalima

```python
for index1, l1 in df.iterrows():
#     print(l1.line)
    if index1==len(df)-1:
        break
    for index2, l2 in df.loc[index1+1:,].iterrows():
#         print(l2)
#         break
        crossing = l1["line"].intersection(l2["line"])
        if crossing.is_empty == False:
            print(f"Line {l1.Name} and line {l2.Name} have crossing at {crossing}")
    
```
